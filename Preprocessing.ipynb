```python
def dataframeinfo(df):
    data_types = df.dtypes.unique()

    max_length = max([len(col) for col in df.columns])
    if max_length > 13:
        max_length = max_length - 13
    else:
        max_length = 1
    print("Total rows: ", df.shape[0])
    print("Total columns: ", df.shape[1])
    print('')
    print(f"#      Column Name      {' ' * max_length} Non Null Count      Null Count      Non Null Percentage      Null Percentage      Data Type      Unique Values")
    print(f"--     -----------      {' ' * max_length} --------------      ----------      -------------------      ---------------      ---------      -------------")

    for i, value in enumerate(df.columns.tolist()):
        null_count = df[value].isnull().sum()
        non_null_Count = df.shape[0]-null_count
        non_null_percentage = (df.shape[0] - null_count) / df.shape[0]
        null_percentage = 1 - non_null_percentage
        data_type = str(df[value].dtype)
        unique_values = df[value].nunique()
        print(f'{i: <6} {value: <{17+max_length}} {non_null_Count: <19} {null_count: <15} {non_null_percentage: <24.2%} {null_percentage: <20.2%} {data_type: <14} {unique_values}')

    print("")
    print('--------------------------')
    print("Datatype Count:")
    for dtype in data_types:
        dtype_count = len(df.select_dtypes(include=[dtype]).columns)
        print(f" - {dtype}: {dtype_count}")
```

```python
columns = df_1.columns
treatskewness = []
for i in columns:
    skw = df_1[i].skew()
    print(f'Skewness of {i} is {skw}')
    plot(df_1, i)
    print('')
    if skw >= 0.5 or skw <= -0.5:
        treatskewness.append(i)
```

```python
treatskewness


# below columns are aysmmetric more than 0.5 and less than -0.5. Therefore we have to treat skewness for these columns
```

```python
# applying log transformation for treating outliers
import numpy as np

df_sqrt = df_1.copy()

print('Skewness after log transformation')
for col in treatskewness:
  df_sqrt[col] = np.sqrt(df_sqrt[col])
  print(f'Skewness for {col} is {df_sqrt[col].skew()}')
```

```python
for col in treatskewness:
  plot(df_sqrt, col)
```

```python

df_outlier = df_sqrt.copy()
numerical_columns = df_outlier.select_dtypes(include=['number']).columns

# Create an empty dictionary to store outlier counts for each column
outlier_row_counts = {}
outlier_counts = {}
outlier_columns = []

for col in numerical_columns:
    # Calculate Q1 and Q3
    Q1 = df_outlier[col].quantile(0.25)
    Q3 = df_outlier[col].quantile(0.75)
    
    # Calculate IQR
    IQR = Q3 - Q1
    
    # Define upper and lower bounds for outliers
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Count outliers in each row
    outlier_row_counts[col] = ((df_outlier[col] < lower_bound) | (df_outlier[col] > upper_bound)).sum()
    outlier_counts[col] = ((df_outlier[col] < lower_bound) | (df_outlier[col] > upper_bound)).sum()


    columns_with_outliers = sum(count > 0 for count in outlier_counts.values())

print("Number of rows affected by outliers in each column:")
for col, count in outlier_row_counts.items():
    if count > 0 :
        print(f"{col}: {count}")
        outlier_columns.append(col)

# print("Number of columns with outliers:", columns_with_outliers)
# print("Outlier counts for each column:", outlier_counts)
```

```python

def outlier(df, column):
    skewness_before = df[column].skew()
    iqr = df[column].quantile(0.75) - df[column].quantile(0.25)
    upper_threshold = df[column].quantile(0.75) + (1.5*iqr)
    lower_threshold = df[column].quantile(0.25) - (1.5*iqr)
    df[column] = df[column].clip(lower_threshold, upper_threshold)
    skewness_after = df[column].skew()

    print(f"Skewness of '{column}' before handling outliers: {skewness_before}")
    print(f"Skewness of '{column}' after handling outliers: {skewness_after}")
    print()


for col in outlier_columns:
    outlier(df_outlier, col)
    plot(df_outlier, col)
```
